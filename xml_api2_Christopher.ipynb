```python
import pandas as pd
import numpy as np
from lxml import etree
import xml.etree.ElementTree as ET
import urllib3
```

```python
status_set = {202, 408, 409, 413, 425, 429, 500, 502, 503, 504, 521, 598, 599}
retries = urllib3.util.retry.Retry(backoff_factor=60, status_forcelist=status_set)
http = urllib3.PoolManager(retries=retries)

parser = etree.XMLParser()
```

```python
base_url = "https://boardgamegeek.com/xmlapi2/thing"
params = {'type' : 'boardgame', 'stats' : 0}
folder = "data/"

nums = np.arange(0,101,100)
```

```python

for idx in np.arange(len(nums)-1):
    rg = str(np.arange(nums[idx],nums[idx+1]))[1:-1].replace(' ',',').replace('\n','')
    fields = {**params, 'id' : rg, 'excludesubtype': 'boardgameexpansion'}
    resp = http.request('GET', base_url, fields=fields)
    print('Reponse status:', resp.status)
    filename = f"games_id_{nums[idx+1]}.xml"
    path = f"{folder}{filename}"
    with open(path, 'wb') as fp:
        fp.write(resp.data)
```

```python
parser = etree.XMLParser(ns_clean=True)
tree   = etree.parse("data/games_id_0.xml", parser)
root = tree.getroot()
for child in root:
    print({x.tag for x in root.findall(child.tag+"/*")})
```
