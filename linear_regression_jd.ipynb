```python
from sql_functions import get_dataframe
import pandas as pd
import Capstone_functions as sf
```

```python
schema = "bgg_data"
main = "ml_boardgame_stats"
subdomain = "subdomain"
unique_subdomain = "unique_subdomain"
kickstarter = "clean_ks_with_bgg_id"


df_main = get_dataframe(f"SELECT * FROM {schema}.{main}")
df_marketplace = sf.avg_price_from_marketplace()
df_sub = get_dataframe(f"SELECT * FROM {schema}.{subdomain}")
df_u_sub = get_dataframe(f"SELECT * FROM {schema}.{unique_subdomain}")
df_ks = get_dataframe(f"SELECT * FROM {schema}.{kickstarter}")
```

```python
df_3 = pd.merge(df_main, df_marketplace, how="inner", on="id")
df_3 = pd.merge(df_3, df_sub, how="left", on="id")
df_3 = pd.merge(df_3, df_u_sub, how="left", on="subdomain_id")
```

```python
df_3 = df_3[df_3["subdomain_id"].notna()]
df_3
```

```python
df_4 = df_3.groupby('id').agg({"subdomain_name":"unique"})
```

```python
df_5 = pd.merge(df_3, df_4, how="inner", on="id")
df_5
```

```python
df_5 = df_5.drop(["subdomain_name_x", "subdomain_id"], axis=1)
df_5
```

```python
df_5 = df_5.drop_duplicates("id")
df_5
```

```python
subdomain_dummy = pd.get_dummies(df_5.subdomain_name_y.apply(pd.Series).stack(), drop_first=True).sum(level=0)
subdomain_dummy
```

```python
df_6 = pd.concat([subdomain_dummy, df_5], axis=1)
df_6 = df_6.drop(["subdomain_name_y"], axis=1)
df_6 = df_6.drop(["max_playtime"], axis=1)
df_6
```

```python
df_ks["id"] = df_ks["bgg_id"]
df = pd.merge(df_6, df_ks, on="id", how="inner")
df
```

```python
df.isnull().sum()
```

```python
df["min_playtime"] = df["min_playtime"].fillna(0)
```

```python
df.columns = df.columns.str.replace(" ", "")
df.columns
```

```python
X = df[["Children's", "Customizable", "Family", "Party", "Strategy", "Thematic", "Wargames", "min_players", "max_players", "min_playtime", "min_age", "wanting", "wishing", "averageweight", "avg_price"]]
```

```python
y = df.usd_pledged
```

```python
# import the statsmodels.api module
import statsmodels.api as sm
X = sm.add_constant(X)
```

```python
# create an OLS model
our_model = sm.OLS(y, X)

# use the data to calculate the intercept and slope
model_results = our_model.fit()

# return the output of the model
model_results.summary() # summary contains eg. 'const' (intercept) and 'slope' of the regression equation.
```
