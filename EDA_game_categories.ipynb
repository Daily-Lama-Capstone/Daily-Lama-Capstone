```python
schema = "bgg_data"


df_categories_with_names = get_dataframe(f"SELECT c.id, c.cat_id, uc.cat_name FROM {schema}.category c INNER JOIN {schema}.unique_category uc ON c.cat_id = uc.cat_id")
```

```python
df_categories_with_names
```

```python
mask = df_categories_with_names['cat_name'] == 'print  play'

df_categories_with_names.loc[mask,'cat_name'] = 'print & play'
```

```python
game_components = ['card game','territory building','miniatures','dice','trivia','word game','print & play','collectible components','real-time']
```

```python
mask = df_categories_with_names['cat_name'].isin(game_components)

df_game_components = df_categories_with_names.loc[mask,:].copy()
```
