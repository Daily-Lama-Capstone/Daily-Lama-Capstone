# LXML Parser for BGG XML API

```python
import pandas as pd
from lxml import etree as et


main_d = [] 
name_d = []
nump_comm_d = []
age_comm_d = []
lang_d_comm = []
b_cat = []
b_mech = []
b_family = []
b_expansion = []
b_implementation = []
b_designer = []
b_artist = []
b_publisher = []
b_statistics = []
b_marketplace = []

parser = et.XMLParser(recover=True)
xtree = et.parse("sample.xml", parser= parser).getroot()


for node in xtree:
    id = node.attrib.get("id")
    type = node.attrib.get("type")
    description = node.find("description").text
    yearpublished = node.find("yearpublished").get("value")
    minplayers = node.find("minplayers").get("value")
    maxplayers = node.find("maxplayers").get("value")
    playingtime = node.find("playingtime").get("value")
    maxplayingtime = node.find("maxplaytime").get("value")
    minplayingtime = node.find("minplaytime").get("value")
    minage = node.find("minage").get("value")
    for node2 in node.findall('name'):
        name_d.append({"id": id ,"name": node2.attrib['value'],"type_name": node2.attrib['type']})
    for node2 in node.findall('link'):
        if node2.attrib['type'] == 'boardgamecategory':
            b_cat.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgamemechanic':
            b_mech.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgamefamily':
            b_family.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgameexpansion':
            b_expansion.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgameimplementation':
            b_implementation.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgamedesigner':
            b_designer.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgameartist':
            b_artist.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
        if node2.attrib['type'] == 'boardgamepublisher':
            b_publisher.append({"id": id ,"cat_id": node2.attrib['id'] ,"cat_name": node2.attrib['value']})
    for node3 in node.findall('poll'):
        if node3.attrib['name'] == "suggested_numplayers":
            for node4 in node3:
                for node5 in node4:
                    if node5.attrib['numvotes'] != "0":
                        nump_comm_d.append({"id": id ,"num_players_recomm": node5.attrib['value'] + " for " + node4.attrib["numplayers"] + " player(s)", "num_votes": node5.attrib['numvotes']})
        if node3.attrib['name'] == 'suggested_playerage':
            for node4 in node3:
                for node5 in node4:
                    if node5.attrib['numvotes'] != "0":
                        age_comm_d.append({"id": id ,"age": node5.attrib['value'], "num_votes": node5.attrib['numvotes']})
        if node3.attrib['name'] == 'language_dependence':
            for node4 in node3:
                for node5 in node4:
                    if node5.attrib['numvotes'] != "0":
                        lang_d_comm.append({"id": id ,"language_dep": node5.attrib['value'], "num_votes": node5.attrib['numvotes']})
    for node2 in node.findall('statistics'):
        for node3 in node2:
                b_statistics.append({"id": id, 
                "average": node3.find("average").get("value"),
                "user_rated": node3.find("usersrated").get("value"),
                "num_owned": node3.find("owned").get("value"),
                "trading": node3.find("trading").get("value"),
                "wanting": node3.find("wanting").get("value"),
                "wishing": node3.find("wishing").get("value"),
                "numcomments": node3.find("numcomments").get("value"),
                "numweights": node3.find("numweights").get("value"),
                "averageweight": node3.find("averageweight").get("value"),
                })
        for node2 in node.findall('marketplacelistings'):
            for node3 in node2:
                    b_marketplace.append({
                    "id": id, 
                    "listdate": node3.find("listdate").get("value"),
                    "price": node3.find("price").get("value"),
                    "currency": node3.find("price").get("currency"),
                    "condition": node3.find("condition").get("value")
                    })
    main_d.append({"id": id ,"type": type ,"description": description, "yearpublished": yearpublished, 
                    "min_players": minplayers,"max_players": maxplayers, "playtime":playingtime, "min_playtime":minplayingtime, 
                    "max_playtime":maxplayingtime, "min_age": minage})
        

df_main = pd.DataFrame(main_d).set_index("id")
df_name = pd.DataFrame(name_d).set_index("id")
df_poll = pd.DataFrame(nump_comm_d).set_index("id")
df_age_comm = pd.DataFrame(age_comm_d).set_index("id")
df_lang_d_comm = pd.DataFrame(lang_d_comm).set_index("id")
df_b_cat = pd.DataFrame(b_cat).set_index("id")
df_b_mech = pd.DataFrame(b_mech).set_index("id")
b_family = pd.DataFrame(b_family).set_index("id")
b_expansion = pd.DataFrame(b_expansion).set_index("id")
b_implementation = pd.DataFrame(b_implementation).set_index("id")
b_designer = pd.DataFrame(b_designer).set_index("id")
b_publisher = pd.DataFrame(b_publisher).set_index("id")
b_statistics = pd.DataFrame(b_statistics).set_index("id")
b_marketplace = pd.DataFrame(b_marketplace).set_index("id")

display(df_main)
display(df_name)
display(df_poll)
display(df_age_comm)
display(df_lang_d_comm)
display(df_b_cat)
display(df_b_mech) 
display(b_family) 
display(b_expansion)
display(b_implementation)
display(b_designer)
display(b_statistics)
display(b_publisher) 
display(b_marketplace) 
```
